```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define vectors
u = np.array([2, 0, 0])
v = np.array([0, 3, 0])
cross = np.cross(u, v)

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Origin
origin = np.array([0, 0, 0])

# Plot u (red), v (blue), cross product (green)
ax.quiver(*origin, *u, color='r', label="u")
ax.quiver(*origin, *v, color='b', label="v")
ax.quiver(*origin, *cross, color='g', label="u × v")

# Set limits
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 7])

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Cross Product Visualization")
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def vector_magnitude(v):
    return np.linalg.norm(v)

def vector_addition(a, b):
    return a + b

def vector_subtraction(a, b):
    return a - b

def vector_dot(a, b):
    return np.dot(a, b)

def interpret_dot(dot_value):
    if dot_value > 0:
        return "Vectors point in a similar direction."
    elif dot_value < 0:
        return "Vectors point in opposite directions."
    else:
        return "Vectors are perpendicular."

def plot_vectors(a, b):
    # Calculate addition and subtraction
    add = a + b
    sub = a - b
    dot = np.dot(a, b)
    interpretation = interpret_dot(dot)

    plt.axhline(0, color="gray", linewidth=0.5)
    plt.axvline(0, color="gray", linewidth=0.5)

    # Plot A (red)
    plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label="A")
    # Plot B (blue)
    plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label="B")
    # Plot A+B (green)
    plt.quiver(0, 0, add[0], add[1], angles='xy', scale_units='xy', scale=1, color='g', label="A+B")
    # Plot A-B (purple)
    plt.quiver(0, 0, sub[0], sub[1], angles='xy', scale_units='xy', scale=1, color='m', label="A-B")

    # Set axis limits
    max_val = int(max(abs(a).max(), abs(b).max(), abs(add).max(), abs(sub).max())) + 2
    plt.xlim(-max_val, max_val)
    plt.ylim(-max_val, max_val)

    # Add labels and legend
    plt.title(f"Vector Playground\nDot product: {dot} â {interpretation}")
    plt.grid(True)
    plt.legend()
    plt.show()


# ---- MAIN PROGRAM ----
print("=== Vector Playground with Plotting ===")

# Input vectors
a = np.array(list(map(float, input("Enter vector A (comma separated): ").split(","))))
b = np.array(list(map(float, input("Enter vector B (comma separated): ").split(","))))

print("\nResults:")
print("Magnitude of A:", vector_magnitude(a))
print("Magnitude of B:", vector_magnitude(b))
print("A + B =", vector_addition(a, b))
print("A - B =", vector_subtraction(a, b))
dot = vector_dot(a, b)
print("Dot product (A Â· B) =", dot)
print("Interpretation:", interpret_dot(dot))

# Plot everything
plot_vectors(a, b)
```
